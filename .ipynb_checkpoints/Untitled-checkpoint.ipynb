```python
import pandas as pd
from datetime import datetime

data = pd.read_csv('sphist.csv')
print(data.dtypes)
print(data.columns)
data['Date'] = pd.to_datetime(data['Date'])
print(data.dtypes)

print(data.head())
data.sort_values('Date',inplace=True,ascending=True)
print(data.head())
```

```python
data["Date"] > datetime(year=2015, month=4, day=1)
```

```python
next(data.iterrows())[1]
```

```python
data['day 5'] = data['Close'].rolling(window=5).mean().shift(periods=1)
data['day 10'] = data['Close'].rolling(window=10).mean().shift(periods=1)
data['day 365'] = data['Close'].rolling(window=250).mean().shift(periods=1)

print(data.iloc[0])
```

```python
data[data["Date"] > datetime(year=1951, month=1, day=2)]
```

```python
new_data = data.dropna(axis=0,inplace=False, how='any')
print(new_data)
```

```python
train = new_data[new_data['Date'] < datetime(year = 2013, month = 1, day=1)]
train_X = train.drop(labels = ['Close','Date'] , axis = 1)
train_Y = train['Close']
print(train_X.tail())

test = new_data[new_data['Date'] >= datetime(year = 2013, month = 1, day = 1)]
test_X = test.drop(labels = ['Close','Date'] , axis = 1)
test_Y = test['Close']
print(test_X.head())
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_X, train_Y)
a0 = lr.intercept_
a1 = lr.coef_

predictions = lr.predict(test_X)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(test_Y,predictions)
print(mae)
```

```python
train_X = train[['day 5','day 10','day 365']]
train_Y = train['Close']
print(train_X.tail())

test_X = test[['day 5','day 10','day 365']]
test_Y = test['Close']
print(test_X.head())
```

```python
data['std 5'] = data['Close'].rolling(window=5).std().shift(periods=1)
data['std 10'] = data['Close'].rolling(window=10).std().shift(periods=1)
data['std 365'] = data['Close'].rolling(window=250).std().shift(periods=1)

data
```

```python
new_data = data.dropna(axis=0,inplace=False, how='any')
print(new_data)

train = new_data[new_data['Date'] < datetime(year = 2013, month = 1, day=1)]
train_X = train[['day 5','day 10','day 365','std 5','std 10','std 365']]
train_Y = train['Close']
print(train_X.tail())

test = new_data[new_data['Date'] >= datetime(year = 2013, month = 1, day = 1)]
test_X = test[['day 5','day 10','day 365','std 5','std 10','std 365']]
test_Y = test['Close']
print(test_X.head())

lr = LinearRegression()
lr.fit(train_X, train_Y)
a0 = lr.intercept_
a1 = lr.coef_

predictions = lr.predict(test_X)
mae = mean_absolute_error(test_Y,predictions)
print(mae)
```
